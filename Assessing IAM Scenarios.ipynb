### Gas Based Electricity Generation

```python
gas_generation = db.filter(variable = "Secondary Energy|Electricity|Gas")
gas_generation
```

### Historical Gas Generation 

```python
gas_generation.filter(year = [2005, 2010, 2015, 2020], region = "World").plot(color = "scenario")
plt.savefig("figures\Gas based electricity generation in IAMs Historical.png", dpi = 300)
plt.show()
```

### Historical Gas Generation and Projections

```python
gas_generation.filter(region = "World").plot(color = "scenario")
plt.savefig("figures\Gas based electricity generation in IAMs Historical and projections.png", dpi = 300)
plt.show()
```

## Gas Share

### Calculate Gas share in electricity generation in scenarios

```python
db.divide(a="Secondary Energy|Electricity|Gas",
                     b = "Secondary Energy|Electricity",
                     name = "Gas share in electricity generation (%)",
                     ignore_units = True,
                     append = True)
```

```python
gas_share = db.filter(variable = "Gas share in electricity generation (%)")
gas_share
```

```python
gas_share.timeseries()
```

### Gas Share Historical

```python
gas_share.filter(year = [2005, 2010, 2015, 2020], region = "World").plot(color = "scenario")
plt.ylabel("Share (0-1)")
plt.savefig("figures\Gas share in electricity generation in IAMs Historical.png", dpi = 300)
plt.show()
```

### Historical and future projections

```python
gas_share.filter(region = "World").plot(color = "scenario")
plt.ylabel("Share (0-1)")
plt.savefig("figures\Gas share in electricity generation in IAMs Historical and projections.png", dpi = 300)
plt.show()
```

## Unabated Gas and Gas with CCS

Gas with CCS
Secondary Energy|Electricity|Gas|w/ CCS


```python
#Unabated gas in electricity generation
db.divide(a="Secondary Energy|Electricity|Gas|w/o CCS",
         b="Secondary Energy|Electricity",
         append = True,
         ignore_units = True,
         name = "Unabated gas share in electricity generation")

#gas with CCS in electricity generation
db.divide(a="Secondary Energy|Electricity|Gas|w/ CCS",
         b="Secondary Energy|Electricity",
         append = True,
         ignore_units = True,
         name = "Gas with CCS share in electricity generation")
```

```python
unabated_gas = db.filter(variable = "Unabated gas share in electricity generation",
                region = "World")

gas_with_CCS = db.filter(variable = "Gas with CCS share in electricity generation",
                region = "World")
unabated_gas
```

```python
fig, ax = plt.subplots()

ax.fill_between(x = [2035, 2040, 2045, 2050],
                 y1 = [0.10, 0.05, 0.025, 0],
                y2 = [0.05, 0, 0, 0],
                alpha = 0.5,
               color = "red",
               label = "CA Benchmarks")

ax.plot(unabated_gas.data["year"],
        unabated_gas.data["value"],
       alpha = 1,
       linewidth = 0.25,
       label = "IAM Scenarios")

iea["Unabated Natural gas share"].plot(ax = ax, color = "darkgreen", label = "IEA Net Zero Scenario 2021")

plt.title("Unabated gas share in global electricity generation")
plt.ylabel("Share (0-1)")
plt.legend()

plt.savefig("figures/Unabated gas share in global electricity generation scenarios comparison",
           dpi = 300)
plt.show()
```
